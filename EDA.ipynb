```python
from dataprep.eda import create_report
import pandas as pd
import os
```

```python
DATA_PATH = r'/mnt/c/Users/chaehyun/Dropbox/Work/PIPET/과제/혈액내과/lpl/Data/WM final.xlsx'
OUTPUT_DIR = r'/mnt/c/Users/chaehyun/Dropbox/Work/PIPET/과제/혈액내과/lpl/EDA'

# 출력 폴더가 없으면 생성
os.makedirs(OUTPUT_DIR, exist_ok=True)
```

```python
all_sheets = pd.read_excel(DATA_PATH, sheet_name=None, engine='openpyxl')
```

```python
import re

for sheet_name, df in all_sheets.items():
    # 빈 시트 스킵
    if df.empty:
        print(f'[SKIP] {sheet_name} — 데이터 없음')
        continue

    # 파일명에 쓸 수 없는 문자 치환
    safe_name = re.sub(r'[\\/:*?"<>|]', '_', sheet_name).strip()
    output_path = os.path.join(OUTPUT_DIR, safe_name)

    print(f'[EDA] {sheet_name} ({df.shape[0]}행 x {df.shape[1]}열) → {safe_name}.html')

    try:
        report = create_report(df)
        report.save(output_path)  # .html 자동 추가
        print(f'  ✓ 저장 완료')
    except Exception as e:
        print(f'  ✗ 오류: {e}')
```
